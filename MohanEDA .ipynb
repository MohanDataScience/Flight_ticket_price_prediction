# Feature Engineering Process

```python
final_df['Date']=final_df["Date_of_Journey"].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
final_df.head(2)
```

```python
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop(['Date_of_Journey'],axis=1,inplace=True)
```

```python
final_df.head(2)
```

```python
final_df['Arrival_Time'].str.split(' ').str[:]
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head(2)
```

```python
final_df['Arrival_hour']=final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_minutes']=final_df['Arrival_Time'].str.split(':').str[1]
final_df.drop(['Arrival_Time'],axis=1,inplace=True)
final_df.head(2)
```

```python
final_df["Arrival_hour"]=final_df["Arrival_hour"].astype(int)
final_df['Arrival_minutes']=final_df['Arrival_minutes'].astype(int)
```

```python
final_df.info()
```

```python
final_df['Dep_hour']=final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_minute']=final_df['Dep_Time'].str.split(":").str[1]
final_df['Dep_hour']=final_df['Dep_minute'].astype(int)
final_df['Dep_minute']=final_df['Dep_minute'].astype(int)
final_df.drop(['Dep_Time'],axis=1,inplace=True)
```

```python
final_df.info()
```

```python
final_df["Total_Stops"].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df["Total_Stops"]=final_df['Total_Stops'].fillna(2)
```

```python
final_df['Total_Stops'].isnull().sum()
```

```python
final_df.info()
```

```python
final_df.head()
```

# Very crucial code

```python
final_df["Duration_hour"]=final_df['Duration'].str.strip(' ').str[0]
```

```python
final_df["Duration_hour"]=final_df['Duration'].str.strip('h').str[0]
```

```python
final_df['Duration_hour']=final_df['Duration_hour'].astype(int)
```

```python
final_df.info()
```

```python
final_df['Duration_minute']=final_df['Duration'].str.strip(' ').str[3:5]
final_df['Duration_minute']
```

```python
final_df['Duration_minute']=final_df["Duration_minute"].replace('',0)
```

```python
final_df['Duration_minute']=final_df['Duration_minute'].replace('5m',5)
```

```python
final_df[final_df['Duration_minute']=='5m']
```

```python
final_df['Duration_minute']=final_df['Duration_minute'].astype(int)
```

```python
final_df["Duration_in_minutes"]=final_df['Duration_hour']*60+final_df['Duration_minute']
```

```python
final_df.head()
```

```python
final_df.drop(['Duration'],axis=1,inplace=True)
```

```python
final_df.info()
```

```python
final_df.drop(['Route'],axis=1,inplace=True)
```

```python
final_df['Additional_Info'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
final_df["Airline"]=labelencoder.fit_transform(final_df['Airline'])
final_df['Source']=labelencoder.fit_transform(final_df['Source'])
final_df['Destination']=labelencoder.fit_transform(final_df['Source'])
final_df['Additional_Info']=labelencoder.fit_transform(final_df['Additional_Info'])
```
